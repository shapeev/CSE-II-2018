```python
# From http://scipy.github.io/devdocs/tutorial/fftpack.html#discrete-sine-transforms

from scipy.fftpack import dct, idct, dst, idst
import matplotlib.pyplot as plt
import numpy as np
N = 100
t = np.linspace(0,20,N)
x = np.exp(-t/3)*np.cos(2*t)
y = dct(x, norm='ortho')
window = np.zeros(N)
window[:20] = 1
yr = idct(y*window, norm='ortho')
sum(abs(x-yr)**2) / sum(abs(x)**2) # 0.0010901402257
```

```python
plt.plot(t, x, '-bx')
plt.plot(t, yr, 'ro')
window = np.zeros(N)
window[:15] = 1
yr = idct(y*window, norm='ortho')
sum(abs(x-yr)**2) / sum(abs(x)**2) # 0.0718818065008
```

```python
plt.plot(t, yr, 'g+')
plt.legend(['x', '$x_{20}$', '$x_{15}$'])
plt.grid()
plt.show()
```

```python
N=9
x = np.linspace(1.0/(N+1),(N+0.0)/(N+1),N)
x
```

```python
amp=dst(0*x+1,norm='ortho') # \hat{f}
```

```python
# blue is the numerical solution, blue is the exact solution
plt.plot(x,idst(amp/((np.pi*np.linspace(1,N,N))**2),norm='ortho'));
plt.plot(x,x*(1-x)/2,'g');
plt.show()
```

```python
N=1000000
t = np.linspace(1.0/(N+1),(N+0.0)/(N+1),N)
amp=dst(0*t+1,norm='ortho')
plt.plot(t,idst(amp/((np.pi*np.linspace(1,N,N))**2),norm='ortho'));
plt.plot(t,t*(1-t)/2,'g');
plt.show()
```
